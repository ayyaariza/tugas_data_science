## 1. *Sorting*

Hal pertama yang kita lakukan ialah mengimport library nya terlebih dahulu

```python
import numpy as np
import pandas as pd
```

kemudian kita baca data CSV yang telah kita sediakan

```python
food = pd.read_csv('food_delivery_datasets.csv')
food
```

setelah itu kita akan mencoba melakukan sorting data secara ascending dan descending untuk dapat melihat urutan data dari yang terendah dan tertinggi.  Kali ini akan dilakukan sorting food_price Dataframe Food_Delivery secara ascending agar dapat mengetahui urutan harga makanan dari yang terendah hingga tertinggi

```python
data1 = food.sort_values(by='food_price', ascending=True)
data1
```

selanjutnya akan dilakukan sorting delivery_fee Dataframe Food secara descending agar dapat mengetahui urutan delivery_fee dari yang tertinggi hingga terendah. Dalam sorting descending dapat ditambahkan parameter ascending=false

```python
data2 = food.sort_values(by='delivery_fee', ascending=False)
data2
```

## 2.*Filtering* 
Dari hasil sortingan diatas kita akan mencoba melakukan filtering untuk menemukan  5 harga makanan terendah dan delivery fee termahal . 
selanjutnya kita ambil data dari dataframe menggunakan rentang kolom dan baris. untuk menemukan  5 harga makanan terendah kita tampilkan data dari dataframe baris 1 sampai 5 dari hasil sorting food_price Dataframe Food_Delivery secara ascending


```python
data1[0:5]
```

Untuk menemukan delivery fee termahal kita tampilkan data dari dataframe baris 0 sampai 1 dari hasil sorting delivery_fee Dataframe Food_Delivery secara descending 

```python
data2[0:1]
```

kemudian kita pilih satu kolom pada Dataframe untuk melihat kolom status saja yang ditampilkan, kita dapat menggunakan []

```python
status_order = food["status"] 
status_order
```

## 3.*Grouping*
Dalam pandas kita juga dapat melakukan pengelompokkan data berdasarkan nama tertentu dan mendapatkan nilainya, entah itu jumlahnya, rata-rata atau perhitungan totalnya. pada dataset Food_Delivery kita dapat melihat  jumlah orderan yang cancelled dan completed 

```python
pd.DataFrame(food.groupby(['status'])['status'].size())
```

## Kesimpulan
Berdasarkan hasil sorting, filtering, dan grouping, dapat diketahui bahwa


1.   5 harga terendah dari makanan yang telah diorder ialah 12900, 14200, 15200, 15700, 16000,
2.   fee delivery termahal ialah 44000
